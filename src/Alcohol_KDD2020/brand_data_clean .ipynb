```python
import copy
import pandas as pd
from urllib.parse import urlparse

def create_new_instance(tagged_user_info, original_data):
    new_instance = copy.deepcopy(original_data)
    tagged_user_id = tagged_user_info['id']
    tagged_user_username = tagged_user_info['username']
    new_instance['influencerId'] = tagged_user_id
    new_instance['influencerUsername'] = tagged_user_username
    return new_instance

def extract_brand_name(instagram_url):
    parsed_url = urlparse(instagram_url)
    path_segments = parsed_url.path.strip('/').split('/')
    return path_segments[0] if path_segments else None

def process_data(input_df):
    new_instances = []

    for index, row in input_df.iterrows():
        tagged_users_info = row['taggedUsers']
        if isinstance(tagged_users_info, list):
            new_instances.extend([create_new_instance(tagged_user_info, row.to_dict()) for tagged_user_info in tagged_users_info])

    new_df = pd.DataFrame(new_instances)
    new_df['brand_name'] = new_df['inputUrl'].apply(extract_brand_name)

    columns_to_drop = ['id','type','inputUrl','shortCode','mentions','url','firstComment','latestComments',
                         'dimensionsHeight','dimensionsWidth','displayUrl','images','timestamp',
                         'childPosts','isSponsored','ownerFullName','ownerUsername','ownerId'
                         ,'videoUrl','alt','productType','isPinned','locationId','taggedUsers',
                         'locationName']
    new_df = new_df.drop(columns=columns_to_drop, axis=1)
    return new_df

json_files = categories
main_df = pd.concat([pd.read_json(f'./brand/{file}') for file in json_files], ignore_index=True)
processed_main_df = process_data(main_df)
processed_main_df = processed_main_df[processed_main_df['brand_name'] != processed_main_df['influencerUsername']]
processed_main_df.reset_index(drop=True, inplace=True)
```
